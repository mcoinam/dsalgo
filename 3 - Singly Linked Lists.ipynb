### Singly Linked Lists :
* collection of __nodes__
* a linear data structure (sequential)
* each node stores a reference to an element of the sequence as well as a reference to the next node of the list
* does not have a predetermined fixed size

```python
class Node(object):
    
    def __init__(self,value):
        self.value = value
        self.nextNode = None
    
```

```python
a = Node(1)
b = Node(2)
c = Node(3)
```

```python
a.nextNode = b
b.nextNode = c
```

```python
a.value
```

```python
a.nextNode.value
```

### Pros :
* constant time insertions and deletions in any position whereas array require O(n) for the same job
* linked lists can expand without specifying their size ahead of time

### Cons :
* accessing an element from a linked list is expensive : O(k) to access the $k^{th}$ element whereas arrays can do this in constant time
